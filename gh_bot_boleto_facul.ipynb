```python
#importações das bibliotecas
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import date
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import time
import urllib
import pyautogui
from PIL import Image
import subprocess
import pyperclip
import smtplib
import email.message

# definindo uma pausa para cada execução do código
pyautogui.PAUSE = 2

# mantém o webdriver do Chrome sempre atualizado
servico = Service(ChromeDriverManager().install())

# abrindo o site
navegador = webdriver.Chrome(service=servico)
navegador.get("link_do_portal_faculdade_financeiro")
time.sleep(5)

# deixando em tela cheia
navegador.maximize_window()

# passando pelo login e senha
time.sleep(5)
id_login = navegador.find_element(By.ID, "User")
id_senha = navegador.find_element(By.ID, "Pass")

id_login.send_keys("login")
id_senha.send_keys("senha")

id_senha.send_keys(Keys.RETURN)
time.sleep(5)

# rolando a tela para baixo para dar o print
time.sleep(3)
navegador.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# print da tela com data de vencimento e valor do boleto
time.sleep(6)
navegador.save_screenshot('print_do_boleto_comvalores.png')
 
navegador.get("https://www.useblackbox.io/signup")
time.sleep(5)

# fazendo o login na plataforma Blackbox para fazer o upload da imagem
login_blackbox = navegador.find_element(By.ID, "w-node-_83ebd445-27bc-da71-4172-773675237666-4a6a3d7d")
login_blackbox.click()
time.sleep(4)

time.sleep(2)
id_email_microsoft = navegador.find_element(By.ID, "i0116")

id_email_microsoft.send_keys("login")
id_email_microsoft.send_keys(Keys.RETURN)
time.sleep(5)

id_senha_microsoft = navegador.find_element(By.ID, "i0118")
id_senha_microsoft.send_keys("senha")
id_senha_microsoft.send_keys(Keys.RETURN)
time.sleep(3)

yes_microsoft = navegador.find_element(By.ID, "acceptButton")
yes_microsoft.click()

time.sleep(3)
while pyautogui.locateOnScreen('blackbox_upload.png', confidence=0.7) is None:
    time.sleep(3)
pyautogui.click('blackbox_upload.png')

time.sleep(2)
for c in range(0,6):
    pyautogui.press('tab')
    
pyautogui.write('print_do_boleto_comvalores.png')
time.sleep(2)

while pyautogui.locateOnScreen('selecionar_boleto2.png', confidence=0.7) is None: 
    time.sleep(1)
time.sleep(6)
pyautogui.click('selecionar_boleto2.png')

pyautogui.press('tab')
pyautogui.press('tab')
pyautogui.press('tab')
pyautogui.press('enter')

chatbot_blackbox = navegador.find_element(By.ID, "chat-input-box")
chatbot_blackbox.send_keys("monte uma mensagem de e-mail muito breve em português brasileiro, onde tudo que estiver entre {} são variáveis que tem que colocar de acordo com as informações da imagem, na seguinte formatação: Assunto (não escreve 'assunto') = Aviso de Vencimento Facul - {nome do aluno} - {data do vencimento do boleto}Olá. Este e-mail foi gerado automaticamente para te lembrar da parcela da faculdade.Valor da Parcela: {valor da parcela} / Data de vencimento: {data de vencimento}{nome do aluno} / RA: {RA do aluno}Segue link do portal do aluno para que possa fazer o pagamento: link_do_portal_faculdade_financeiroAtenciosamente, Splash Brothers Companhia Limitada ©")
chatbot_blackbox.send_keys(Keys.RETURN)

time.sleep(10)
edit_text = navegador.find_element(By.CSS_SELECTOR, "body > div.flex.flex-col.min-h-screen > main > div > div.group.w-full.overflow-auto.pl-0.animate-in.duration-300.ease-in-out.peer-\[\[data-state\=open\]\]\:lg\:pl-\[250px\].peer-\[\[data-state\=open\]\]\:xl\:pl-\[300px\] > div > div > div.pt-4.blackbox-chat-area > div > div:nth-child(2) > div.relative.mx-auto.max-w-2xl.px-4 > div > div.ml-4.flex-1.space-y-2.overflow-hidden.px-1 > div.prose.break-words.dark\:prose-invert.prose-p\:leading-relaxed.prose-pre\:p-0.fix-max-with-100")
print(edit_text.text)

def enviar_email():
    corpo_email = f"<p>{edit_text.text[27:]}</p>"
    
    msg = email.message.Message()
    msg['Subject'] = "Boleto da Faculdade está próximo de vencer!"
    msg['From'] = 'email_remetente'
    msg['To'] = 'email_que_recebera'
    password = 'senhaespecifica_de_logindogmail' #nas novas versões do gmail, é preciso gerar uma senha específica para que conseguimos automatizar o envio
    msg.add_header("Content-Type", 'text/html')
    msg.set_payload(corpo_email)
    
    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    pyautogui.alert('Email enviado com sucesso!')
    
enviar_email()
```
